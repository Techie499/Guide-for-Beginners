```python
import matplotlib.pyplot as plt
```

# Reading Data

```python
with open("Data/Goals.txt","r") as goalData:
    homeTeamLine = goalData.readline()
    homeTeamLine = homeTeamLine.strip(" \n")
    homeTeamLine = homeTeamLine.split(" ")

    HomeTeamGoals = [int(x) for x in homeTeamLine]
    
    awayTeamLine = goalData.readline()
    awayTeamLine = awayTeamLine.strip(" \n")
    awayTeamLine = awayTeamLine.split(" ")

    AwayTeamGoals = [int(x) for x in awayTeamLine]
```

# Histogram

```python
x=[]
for i in range(1,len(HomeTeamGoals)+1):
    x.append(i)
```

```python
fig2 = plt.figure("Second",figsize = (6,6))
ax1 = fig2.add_subplot(1,1,1)

ax1.hist(x = HomeTeamGoals, 
         bins = 7, #int / sequence 
         align = "mid", # left, mid, right
         #range = (1,3), # (min,max)
         color = "red",
         cumulative= False # True, False
)

plt.show()
```

```python
fig2 = plt.figure("Second",figsize = (6,6))
ax1 = fig2.add_subplot(1,1,1)

ax1.hist(x = (HomeTeamGoals, AwayTeamGoals), 
         bins = 7, #int / sequence 
         align = "mid", # left, mid, right
         #range = (1,3), # (min,max)
         color = ("red","blue"),
         cumulative= False # True, False
)

plt.show()
```

```python
fig2 = plt.figure("Second",figsize = (6,6))
ax1 = fig2.add_subplot(1,1,1)

N, bins, patches = ax1.hist(x = (HomeTeamGoals, AwayTeamGoals), 
         bins = 7, #int / sequence 
         align = "mid", # left, mid, right
         #range = (1,3), # (min,max)
         color = ("red","blue"),
         cumulative= False # True, False
)

patches[1][2].set_facecolor("green")
patches[0][3].set_width(0.5)

plt.show()
```

# Bar Graph

```python
avgGoalsScoredHome = sum(HomeTeamGoals)/len(HomeTeamGoals)
avgGoalsScoredAway = sum(AwayTeamGoals)/len(AwayTeamGoals)
```

```python
fig2 = plt.figure("Second",figsize = (6,6))
ax1 = fig2.add_subplot(1,1,1)

ax1.bar( x = [1,2],
        height= (avgGoalsScoredHome,avgGoalsScoredAway),
        width= 0.35,
        align= "center",
        color = ["red","blue"],
        bottom = [0,0],
        )

plt.xticks([1,2],["Home Team","Away Team"])

ax1.set_ylim(0,2)
ax1.set_xlim(0.5,2.5)


plt.show()
```

```python
fig2 = plt.figure("Second",figsize = (6,6))
ax1 = fig2.add_subplot(1,1,1)

ax1.bar( x = [1,1],
        height= (avgGoalsScoredHome,avgGoalsScoredAway),
        width= 0.15,
        align= "center",
        color = ["red","blue"],
        bottom = [0,avgGoalsScoredHome],
        xerr = 0.15,
        #xerr = [0.15,0.4],
        #xerr = [[0.15,0.4],[0.4,0.15]],
        yerr = 0.15,
        #yerr = [0.15,0.4],
        #yerr = [[0.15,0.4],[0.4,0.15]],
        ecolor = "orange",
        )

plt.xticks([1,2],["Home Team","Away Team"])

ax1.set_ylim(0,4)
ax1.set_xlim(0.5,2.5)


plt.show()
```

```python
fig3 = plt.figure("Third", figsize = (6,6))
ax1 = fig3.add_subplot(1,1,1)

ax1.boxplot(x = [HomeTeamGoals,AwayTeamGoals],
          #  positions= [1,5],
          sym = "^", #"o", "s", "d", "v", "^", "<", ">", "p", "h", "+", "x", "D", "|", "_"
          whis  = [10,90],
          #widths = 0.5
          notch= True,
          conf_intervals = [[0.9,1.1],[0.2,1.9]],
)

ax1.set_ylim(0,7)
plt.show()
```

# Pie Charts

```python
totalHomeGoalsScored = sum(HomeTeamGoals)
totalAwayGoalsScored = sum(AwayTeamGoals)

fig2 = plt.figure("Second",figsize = (6,6))
ax1 = fig2.add_subplot(1,1,1)

ax1.pie(x = [totalHomeGoalsScored,totalAwayGoalsScored],
        labels = ["Home Team","Away Team"],
        labeldistance=0.5,
        colors= ["red","green"],
        explode= [0,0.1],
        shadow= True,
        #radius= 1,
        #center= [0,0],
        )

plt.show()
```

# 2 Dimensional Histogram

```python
fig2 = plt.figure("Second",figsize = (6,6))
ax1 = fig2.add_subplot(1,1,1)

plt.hist2d(x = HomeTeamGoals,
           y = AwayTeamGoals,
           bins= [range(8),range(7)]
           )#[xbins,ybins]

plt.colorbar()

plt.show()
```
