# Reading Data

```python
with open("Data/Goals.txt","r") as goalData:
    homeTeamLine = goalData.readline()
    homeTeamLine = homeTeamLine.strip(" \n")
    homeTeamLine = homeTeamLine.split(" ")

    HomeTeamGoals = [int(x) for x in homeTeamLine]
    
    awayTeamLine = goalData.readline()
    awayTeamLine = awayTeamLine.strip(" \n")
    awayTeamLine = awayTeamLine.split(" ")

    AwayTeamGoals = [int(x) for x in awayTeamLine]
```

```python
def distFromZero(htg,atg):
    return math.sqrt(htg**2+atg**2)
```

# Reading a TXT File

```python
def readTxt():
    f = open("Data/Sheet1E1.txt","r")
    data = {}
        
    for line in f:
        part1 = line.strip("\n").split(" ")
        dataName = part1[0]
        restData = part1[1:]

        finalData = []
        for x in restData:
            finalData.append(int(x))
        data[dataName] = finalData
            
    f.close()
    
    return data
```

# Reading a CSV File

```python
def readCSV():
    f = open("Data/Sheet1E1.csv","r")  
    data = {}
    
    firstLine = f.readline().strip("\n").split(",")[1:]

    for name in firstLine:
        data[name] = []    
    
    for line in f:
        processedLine = line.strip("\n").split(",")[1:]
        
        for index in range(len(processedLine)):
            name = firstLine[index]
            dataPoint = int(processedLine[index])
            data[name].append(dataPoint)
        
    return data
```

```python
data = readCSV()
HPAB=[]
print(data.keys())
```
