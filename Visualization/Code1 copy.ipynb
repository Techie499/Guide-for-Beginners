```python
import matplotlib.pyplot as plt
```

# Reading Data

```python
with open("Data/Goals.txt","r") as goalData:
    homeTeamLine = goalData.readline()
    homeTeamLine = homeTeamLine.strip(" \n")
    homeTeamLine = homeTeamLine.split(" ")

    HomeTeamGoals = [int(x) for x in homeTeamLine]
    
    awayTeamLine = goalData.readline()
    awayTeamLine = awayTeamLine.strip(" \n")
    awayTeamLine = awayTeamLine.split(" ")

    AwayTeamGoals = [int(x) for x in awayTeamLine]
```

# Histogram

```python
x=[]
for i in range(1,len(HomeTeamGoals)+1):
    x.append(i)
```

```python
fig2 = plt.figure("Second",figsize = (6,6))
ax1 = fig2.add_subplot(1,1,1)

ax1.hist(x = HomeTeamGoals, 
         bins = 7, #int / sequence 
         align = "mid", # left, mid, right
         #range = (1,3), # (min,max)
         color = "red",
         cumulative= False # True, False
)

plt.show()
```

```python
fig2 = plt.figure("Second",figsize = (6,6))
ax1 = fig2.add_subplot(1,1,1)

ax1.hist(x = (HomeTeamGoals, AwayTeamGoals), 
         bins = 7, #int / sequence 
         align = "mid", # left, mid, right
         #range = (1,3), # (min,max)
         color = ("red","blue"),
         cumulative= False # True, False
)

plt.show()
```

```python
fig2 = plt.figure("Second",figsize = (6,6))
ax1 = fig2.add_subplot(1,1,1)

N, bins, patches = ax1.hist(x = (HomeTeamGoals, AwayTeamGoals), 
         bins = 7, #int / sequence 
         align = "mid", # left, mid, right
         #range = (1,3), # (min,max)
         color = ("red","blue"),
         cumulative= False # True, False
)

patches[1][2].set_facecolor("green")
patches[0][3].set_width(0.5)

plt.show()
```

```python
from matplotlib.patches import Rectangle
```

```python
redRectangle = Rectangle((0,0),1,1,color = "red") #(x,y), width, height, facecolor
blueRectangle = Rectangle((0,0),1,1,color = "blue") #(x,y), width, height, facecolor
```

```python
labels = ["Home Team Goals","Away Team Goals"]
Boxes = [redRectangle,blueRectangle]

plt.legend(Boxes,labels)
```

# Bar Graph

```python
avgGoalsScoredHome = sum(HomeTeamGoals)/len(HomeTeamGoals)
avgGoalsScoredAway = sum(AwayTeamGoals)/len(AwayTeamGoals)
```

```python
fig2 = plt.figure("Second",figsize = (6,6))
ax1 = fig2.add_subplot(1,1,1)

ax1.bar( x = [1,2],
        height= (avgGoalsScoredHome,avgGoalsScoredAway),
        width= 0.35,
        align= "center",
        color = ["red","blue"],
        bottom = [0,0],
        )

plt.xticks([1,2],["Home Team","Away Team"])

ax1.set_ylim(0,2)
ax1.set_xlim(0.5,2.5)


plt.show()
```
